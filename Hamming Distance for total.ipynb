```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gv=[]
iv=[]
count=150000
with open('02463_L.txt') as file:
    for i in file:
        c1,c2,v=i.split()
        v=float(v)
        if v!=0:
            v=int(v*100)/100
        c11,c12=c1.split('d')
        c21,c22=c2.split('d')
        if c11==c21:
            gv.append(v)
        else:
            iv.append(v)
            count-=1
        if count==0:
            break
list=['04233_L.txt','04252_L.txt','04261_L.txt','04267_L.txt','04327_L.txt','04385_L.txt','04394_L.txt']
for j in list:
    with open(j) as file:
        for i in file:
            c1,c2,v=i.split()
            v=float(v)
            if v!=0:
                v=int(v*100)/100
            c11,c12=c1.split('d')
            c21,c22=c2.split('d')
            if c11==c21:
                gv.append(v)
gvl={}
ivl={}
for i in gv:
    if i in gvl:
        gvl[i]+=1
    else:
        gvl[i]=1
for i in iv:
    if i in ivl:
        ivl[i]+=1
    else:
        ivl[i]=1
```

```python
gvname=[]
gvvalue=[]
ivname=[]
ivvalue=[]
gvls={}
for key in sorted(gvl):
    gvls[key]=gvl[key]
ivls={}
for key in sorted(ivl):
    ivls[key]=ivl[key]
for i in gvls:
    gvname.append(i)
    gvvalue.append(gvls[i])
for i in ivls:
    ivname.append(i)
    ivvalue.append(ivls[i])
```

```python
gen={'gen':gvname,'count':gvvalue}
imp={'imp':ivname,'count':ivvalue}
gdf=pd.DataFrame(gen)
idf=pd.DataFrame(imp)
```

```python
plt.plot('gen','count',data=gdf)
plt.plot('imp','count',data=idf)
plt.plot([0.45,0.45],[0,30000])
plt.annotate('Genuine',xy=(0.1,15000),color='blue')
plt.annotate('Impostors',xy=(0.55,20000),color='orange')
plt.annotate('Threshold',xy=(0.34,22500),color='green')
plt.annotate('Threshold = 0.45',xy=(0.7,10000))
plt.xlim(0,0.7)
plt.ylim(0,28000)
plt.grid(True)
plt.title('The Himming Distance of different Iris Code')
plt.xlabel('Himming Distance')
plt.ylabel('Count')
```
